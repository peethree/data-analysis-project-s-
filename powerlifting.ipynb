# Correlation between age, bodyweight and max lifts (without equipment) 

```python
male_df = df[(df['Sex']=='M') & (df['Equipment']=='Raw')]
```

```python
female_df = df[(df['Sex']=='F') & (df['Equipment']=='Raw')]
```

```python
# remove the negative values from the df shown in the graphs above
# filtered for ages between 18 and 30

filtered_male_df = male_df[
    (male_df['Age'].between(18,30)) &
    (male_df['BodyweightKg'] >= 0) &
    (male_df['Best3SquatKg'] >= 0) &
    (male_df['Best3BenchKg'] >= 0) &
    (male_df['Best3DeadliftKg'] >= 0)
]
```

```python
filtered_female_df = female_df[
    (female_df['Age'].between(18,30)) &
    (female_df['BodyweightKg'] >= 0) &
    (female_df['Best3SquatKg'] >= 0) &
    (female_df['Best3BenchKg'] >= 0) &
    (female_df['Best3DeadliftKg'] >= 0)
]
```

```python
# men aged 18-30
df_corr_male = filtered_male_df[['Age','BodyweightKg','Best3SquatKg','Best3BenchKg','Best3DeadliftKg']].dropna().corr()
df_corr_male
```

```python
# correlations for women aged 18-30
df_corr_female = filtered_female_df[['Age','BodyweightKg','Best3SquatKg','Best3BenchKg','Best3DeadliftKg']].dropna().corr()
df_corr_female
```

# Correlation heatmap (males)

```python
sns.heatmap(df_corr_male, annot=True)
```

# Correlation heatmap (females)

```python
sns.heatmap(df_corr_female, annot=True)
```

```python
filtered_male_df_all_ages = male_df[
    (male_df['Age'] >= 0) &
    (male_df['BodyweightKg'] >= 0) &
    (male_df['Best3SquatKg'] >= 0) &
    (male_df['Best3BenchKg'] >= 0) &
    (male_df['Best3DeadliftKg'] >= 0)
]
```

```python
filtered_female_df_all_ages = female_df[
    (female_df['Age'] >= 0) &
    (female_df['BodyweightKg'] >= 0) &
    (female_df['Best3SquatKg'] >= 0) &
    (female_df['Best3BenchKg'] >= 0) &
    (female_df['Best3DeadliftKg'] >= 0)
]
```

```python
df_corr_male_all_ages = filtered_male_df_all_ages[['Age','BodyweightKg','Best3SquatKg','Best3BenchKg','Best3DeadliftKg']].dropna().corr()
df_corr_male_all_ages
```

```python
df_corr_female_all_ages = filtered_female_df_all_ages[['Age','BodyweightKg','Best3SquatKg','Best3BenchKg','Best3DeadliftKg']].dropna().corr()
df_corr_female_all_ages
```

```python
sns.heatmap(df_corr_male_all_ages, annot=True)
```

```python
sns.heatmap(df_corr_female_all_ages, annot=True)
```

```python
# calculate the difference in correlation between the matrices for age 18-30 and all ages for men and women

correlation_matrix_1 = df_corr_male  
correlation_matrix_2 = df_corr_male_all_ages 

assert correlation_matrix_1.shape == correlation_matrix_2.shape, "Matrices must have the same shape"

correlation_difference = np.abs(correlation_matrix_1.values - correlation_matrix_2.values)

overall_difference = np.sum(correlation_difference)

print(f"Overall Difference in Correlation: {overall_difference}")
```

```python
correlation_matrix_1 = df_corr_female  
correlation_matrix_2 = df_corr_female_all_ages

assert correlation_matrix_1.shape == correlation_matrix_2.shape, "Matrices must have the same shape"

correlation_difference = np.abs(correlation_matrix_1.values - correlation_matrix_2.values)

overall_difference = np.sum(correlation_difference)

print(f"Overall Difference in Correlation: {overall_difference}")
```
